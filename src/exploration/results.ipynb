```python
results = hp.read_smth_results('../runs/dummy/smth/tadn_smth_100/results_train_0.json')
for rank in range(1, 4):
    results = results.combine_first(hp.read_smth_results(f'../runs/dummy/smth/tadn_smth_100/results_train_{rank}.json'))
cols = ['top1_conf', 'top1_pred', 'top3_conf_1', 'top3_conf_2', 'top3_conf_3', 'top3_pred_1', 'top3_pred_2', 'top3_pred_3']
for col in cols:
    assert len(results[results[col].isnull()]) == 0, f'{col}, {len(results[results[col].isnull()])}'
results.to_json('../runs/dummy/smth/tadn_smth_100/results_train.json', orient='records')
results
```
